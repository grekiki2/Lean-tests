```python
import matplotlib.pyplot as plt
import numpy as np
import time
import subprocess
from functools import cache

plt.rcParams["figure.figsize"]= (14, 10)
```

```python
# @cache
def run_dump(n, p, col):
    start_time = time.time()
    result = subprocess.run(['../.lake/build/bin/dump', str(n), str(p), str(col)], stdout=subprocess.PIPE)
    end_time = time.time()
    elapsed_time = end_time - start_time
    output = result.stdout.decode('utf-8').split('\n')[1]
    return elapsed_time, output
```

```python
col = 2
xs = []
ls_ys = []
for p in range(0, 100, 25):
    print(p)
    ys = []
    for n in range(100):
        print(n)
        t, output = run_dump(n, p, col)
        if p==0:
            xs.append(n)
        ys.append(t)
        if t>1:
            break
    ls_ys.append(ys)
```

```python
for ls in ls_ys:
    plt.plot(range(len(ls)), ls, 'o-')
plt.grid(axis='y', linestyle='--', alpha=1.0)
plt.xlabel('Število vozlišč', fontsize=20)
plt.ylabel('Čas izvajanja [s]', fontsize=20)
plt.title('Časi računanja 2 barvljivosti', fontsize=20)
plt.legend([f'{p} % povezanost' for p in range(0, 101, 25)], fontsize=20)
plt.tight_layout()
plt.savefig('2barvFast.png')
```
